```python
def get_content_value(data):
    if data.find("li"):
        return [li.get_text(" ",strip=True).replace("\xa0"," ") for li in data.find_all("li")]
    elif data.find("br"):
        return [text for text in data.stripped_strings]
    else:
        return data.get_text(" ",strip=True).replace("\xa0"," ")


# def clean_tag(soup):
#     for tag in soup.find_all("sup","span"):
#         tag.decompose()
#     print(soup)

def get_info(url):
    r = requests.get(url)
    soup= BeautifulSoup(r.content)
    #clean_tag(soup)
    #soup.sup.decompose()
    #print(soup)
    box= soup.find(class_="infobox vevent")
    row = box.find_all("tr")
    #print(row)
    movie_info={}
    
#     clean_tag(soup)
    soup.sup.decompose()
    for index,r in enumerate(row):
        #soup.sup.decompose()
#         while r.find_all("sup"): # to remove references [1] and 22 May 2008 ( 2008-05-22 )
#             soup.sup.decompose()
        for s in soup.select('sup'):
            s.extract()
#         for s in soup.select('span'):
#             s.extract()
        if index==0:
            movie_info['title']=r.find("th").get_text()
        #elif  index==1 or index == 5:
        #    continue
        elif r.find("th")==None: 
            continue
        else:
            key =r.find("th").get_text(" ",strip =True)
            value=get_content_value(r.find("td"))
            movie_info[key]=value

    return movie_info
```

```python
from datetime import datetime

def date_to_datetime(x):
    if x=='NA':
        return None
    if isinstance(x,list):
        x = x[0]
#     if(re.search(f,x).group()):
        #print(re.search(f,x).group())
    if (re.match(f, x)):
        r = re.match(f, x).group()
    else:
        r= x
    fmts = ["%d %B %Y"]
    for  fmt in fmts:
        try:
            return datetime.strptime(r,fmt)
        except:
            pass
    return x

for m in movies_list:
    m['Release date(Datetime)']=date_to_datetime(m.get('Release date', 'NA'))
```

```python
import requests
import urllib

def get_imbd(title):
    base_url ="http://www.omdbapi.com/?"
    #parameters = {"apikey": os.environ['OMBD_key']}
    parameters = {"apikey":"f1373220",'t':title}
    param = urllib.parse.urlencode(parameters)
    full_url = base_url + param
    return requests.get(full_url).json()
def get_rottamoto(info):
    ratings=info.get('Ratings',[])#if it doesnt return anything then give empty list
    for r in ratings:
        #print(r)
        if r['Source'] == 'Rotten Tomatoes':
            return r['Value']
    return None
imdb_info= get_imbd('Baahubali 2: The Conclusion')
get_rottamoto(imdb_info)
```
